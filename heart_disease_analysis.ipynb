```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv('data.csv')
```

```python
heart_data.head()
```

```python
heart_data.tail()
```

```python
heart_data.describe()
```

NUMBER OF ENTRIES AND COLUMNS

```python
heart_data.shape
```

```python
heart_data.isnull().sum()
```

```python
#checking the distribution of the target variable
heart_data["target"].value_counts()
```

1 -> defective heart 
2 -> normal heart

splitting the features and target


```python
x = heart_data.drop(columns = 'target',axis=1)
y = heart_data['target']
```

```python
print(x)
```

```python
#stratify spilt dat in the same proportion as orignal data,
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 , stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

MODEL TRAINING


LOGISTIC REGRESSION

```python
model = LogisticRegression()
```

```python
# training the logistic regression mode; with training data
model.fit(x_train,y_train)
```

Model evaluvation 

Accuracy score

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training data  : ',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Test data  : ',test_data_accuracy)
```

Building a predictive System 

```python
##### input values as features
features = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach","exang", "oldpeak", "slope", "ca", "thal"]
prompts = [
    "Please enter your age in years (e.g., 52): ",
    "Please enter your gender:\n\t1 for male or 0 for female: ",
    "Please select your chest pain type:\n\t0: No chest pain\n\t1: Typical angina\n\t2: Atypical angina\n\t3: Non-anginal pain\n(Enter 0, 1, 2, or 3): ",
    "Please enter your resting blood pressure in mmHg (e.g., 125): ",
    "Please enter your serum cholesterol level in mg/dl (e.g., 212): ",
    "Please enter 1 if your fasting blood sugar is greater than 120 mg/dl, or 0 if it is less than or equal to 120 mg/dl:\n ",
    "Please select the results of your resting electrocardiogram (ECG):\n\t0: Normal\n\t1: Having ST-T wave abnormality\n\t2: Showing probable or definite left ventricular hypertrophy (LVH)\n(Enter 0, 1, or 2):\n ",
    "Please enter the maximum heart rate you achieved during exercise (e.g., 168):\n ",
    "Please enter 1 if you have exercise-induced angina (chest pain), or 0 if you do not:\n ",
    "Please enter the depression induced by exercise relative to rest\n (in units of ST depression, e.g., 1.0): \n\t",
    "Please select the slope of your peak exercise ST segment:\n\t0: Upsloping\n\t1: Flat\n\t2: Downsloping\n(Enter 0, 1, or 2): \t",
    "Please enter the number of major vessels colored by fluoroscopy:\n\t0: No vessels colored\n\t1: 1 vessel colored\n\t2: 2 vessels colored\n\t3: 3 vessels colored\n(Enter 0, 1, 2, or 3): \n",
    "Please enter your thalassemia type:\n\t3: Normal\n\t6: Fixed defect\n\t7: Reversable defect\n(Enter 3, 6, or 7): \n"
]
# data collection 
user_input = []
for prompt in prompts:
    value = input(prompt)
     # Determine if the value should be converted to int or float
    user_input.append(float(value) if '.' in value else int(value))
input_data = tuple(user_input)
#change iput data to numpy array
input_data_as_numpy_array=np.asarray(input_data)
# reshape the numpyarray as we predict for only one data arrray 
in_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
input_data_df = pd.DataFrame(in_data_reshaped,columns=features)
prediction = model.predict(input_data_df)
# print(prediction) 
if (prediction[0]== 0):
    print("\n The person does not have heart disease \n ")
else:
    print("\n The person has heart disease or has very high risk \n")
```
